```python
with rasterio.open(r"//DAta /OUTPUT/nepallc.tif", 'w',
                   driver= ds.driver,
                   height= ds.height,
                   width= ds.width,
                   count= ds.count,
                   csr= ds.crs,
                   transform= ds.transform,
                   dtype= data.dtype,
                   )as dst:
               dst.write(data)
               show(ds, title="Land Cover Map of Nepal")





```
